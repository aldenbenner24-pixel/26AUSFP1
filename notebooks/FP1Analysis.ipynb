```python
distance_common = np.linspace(
    LEClap_fast["Distance"].min(),
    LEClap_fast["Distance"].max(),
    3000
)   
time_LEC_interp = np.interp(
    distance_common,
    LEClap_fast["Distance"],
    LEClap_fast["Time"]
)
time_VER_interp = np.interp(
    distance_common,
    VERlap_fast["Distance"],
    VERlap_fast["Seconds"]
)
time_LIN_interp = np.interp(
    distance_common,
    LINlap_fast["Distance"],
    LINlap_fast["Seconds"]
)
time_PIA_interp = np.interp(
    distance_common,
    PIAlap_fast["Distance"],
    PIAlap_fast["Seconds"]
)
time_RUS_interp = np.interp(
    distance_common,
    RUSlap_fast["Distance"],
    RUSlap_fast["Seconds"]
)
time_HUL_interp = np.interp(
    distance_common,
    HULlap_fast["Distance"],
    HULlap_fast["Seconds"]
)
time_OCO_interp = np.interp(
    distance_common,
    OCOlap_fast["Distance"],
    OCOlap_fast["Seconds"]
)
time_SAI_interp = np.interp(
    distance_common,
    SAIlap_fast["Distance"],
    SAIlap_fast["Seconds"]
)
time_COL_interp = np.interp(
    distance_common,
    COLlap_fast["Distance"],
    COLlap_fast["Seconds"]
)
time_BOT_interp = np.interp(
    distance_common,
    BOTlap_fast["Distance"],
    BOTlap_fast["Seconds"]
)
time_STR_interp = np.interp(
    distance_common,
    STRlap_fast["Distance"],
    STRlap_fast["Seconds"]
) #Find the times that correspond to the new common distances

VERdelta_time = time_VER_interp - time_LEC_interp
LINdelta_time = time_LIN_interp - time_LEC_interp
PIAdelta_time = time_PIA_interp - time_LEC_interp
RUSdelta_time = time_RUS_interp - time_LEC_interp
HULdelta_time = time_HUL_interp - time_LEC_interp
OCOdelta_time = time_OCO_interp - time_LEC_interp
SAIdelta_time = time_SAI_interp - time_LEC_interp
COLdelta_time = time_COL_interp - time_LEC_interp
BOTdelta_time = time_BOT_interp - time_LEC_interp
STRdelta_time = time_STR_interp - time_LEC_interp #Find the delta times for each team's fastest lap

fig, ax1 = plt.subplots(figsize = (20,15))
ax1.plot(distance_common, VERdelta_time, label = 'VER', color = "darkblue")
ax1.plot(distance_common, LINdelta_time, label = 'LIN', color = "deepskyblue")
ax1.plot(distance_common, PIAdelta_time, label = 'PIA', color = "orange")
ax1.plot(distance_common, RUSdelta_time, label = 'RUS', color = "turquoise")
ax1.plot(distance_common, HULdelta_time, label = 'HUL', color = "maroon")
ax1.plot(distance_common, OCOdelta_time, label = 'OCO', color = "lightgray")
ax1.plot(distance_common, SAIdelta_time, label = 'SAI', color = "royalblue")
ax1.plot(distance_common, COLdelta_time, label = 'COL', color = "pink")
ax1.plot(distance_common, BOTdelta_time, label = 'BOT', color = "black")
ax1.plot(distance_common, STRdelta_time, label = 'STR', color = "mediumseagreen")
ax1.set_xlabel("Distance (m)")
ax1.set_ylabel("Delta Time (s)")
ax1.grid(True)
ax1.legend()

ax2 = ax1.twinx()
ax2.plot(LEClap_fast["Distance"], LEClap_fast["Speed"], label = "LEC", color = "red")
ax2.set_xlabel("Speed (km/h")

plt.title("Delta Time vs Distance with Leclerc Reference Lap")
plt.show #Plot the delta time vs the distance 
```
